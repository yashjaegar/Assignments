```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample text
text = ["Natural Language Processing is fascinating. Language is a part of Artificial Intelligence."]

# Initialize the CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the text
word_count = vectorizer.fit_transform(text)

# Convert the result to a dictionary with raw counts
raw_counts = dict(zip(vectorizer.get_feature_names_out(), word_count.toarray()[0]))

# Calculate the total number of words in the corpus
total_words = sum(raw_counts.values())

# Calculate normalized term frequency
term_frequency = {word: count / total_words for word, count in raw_counts.items()}

# Print the normalized term frequency
print(term_frequency)
```
